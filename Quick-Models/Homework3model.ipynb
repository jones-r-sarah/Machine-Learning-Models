```python
import pandas as pd

data = pd.read_csv('C:/Users/jones/Downloads/Advertising.csv', index_col=0)
#first 5 rows, data.tail() is the last 5 rows, data.shape is the dataframe (rows,columns)
data.head() 
```

```python
#python list os featur names (columns)
feature_cols = ['TV', 'radio', 'newspaper']

#selecting a subset of the original datafram #can combine next line and previous x= data[['TV','Radio','Newspaper']]
x = data[feature_cols]

x.head()
```

```python
#select a series from the dataframe
y = data['sales']

y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
#print the intercept and coefficients
print(linreg.intercept_)
print(linreg.coef_)

#pair the feature names with the coefficients
list(zip(feature_cols, linreg.coef_))
```

```python
#make predictions on the testing set
y_pred = linreg.predict(x_test)
```

```python
#define values: true and predicted response
true = [100, 50, 30, 20]
pred = [90,50,50,30]
from sklearn import metrics
import numpy as np
print(np.sqrt(metrics.mean_squared_error(true, pred)))
```
